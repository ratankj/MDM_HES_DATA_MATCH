```python
INFRA_SB['InstallationDatetime'] = pd.to_datetime(INFRA_SB['InstallationDatetime'])
INFRA_SB['EntryDatetime'] = pd.to_datetime(INFRA_SB['EntryDatetime'])
INFRA_SB['LastS05'] = pd.to_datetime(INFRA_SB['LastS05'])
INFRA_SB['LastS02'] = pd.to_datetime(INFRA_SB['LastS02'])
INFRA_SB['LastS04'] = pd.to_datetime(INFRA_SB['LastS04'])


#--------------------------------------------------------------------------------------------------


INFRA_NB['InstallationDatetime'] = pd.to_datetime(INFRA_NB['InstallationDatetime'])
INFRA_NB['EntryDatetime'] = pd.to_datetime(INFRA_NB['EntryDatetime'])
INFRA_NB['LastS05'] = pd.to_datetime(INFRA_NB['LastS05'])
INFRA_NB['LastS02'] = pd.to_datetime(INFRA_NB['LastS02'])
INFRA_NB['LastS04'] = pd.to_datetime(INFRA_NB['LastS04'])
```

```python
INFRA_SB_NOT_NULL = INFRA_SB[~INFRA_SB['LastS05'].isnull()]
INFRA_NB_NOT_NULL = INFRA_NB[~INFRA_NB['LastS05'].isnull()]
```

```python
# Calculate the date 7 days ago
seven_days_ago = datetime.now() - timedelta(days=7)


# Filter the DataFrame to keep only rows before the last 7 days
INFRA_NB_COMM_DATA = INFRA_NB_NOT_NULL[INFRA_NB_NOT_NULL['LastS05'] > seven_days_ago]
INFRA_SB_COMM_DATA = INFRA_SB_NOT_NULL[INFRA_SB_NOT_NULL['LastS05'] > seven_days_ago]


```
